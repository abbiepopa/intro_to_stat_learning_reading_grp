For correlation (and other "traditional" stats) I tend toward scipy (though note the output is both the r and the p)

```python
from scipy import stats

x = np.random.normal(size = 50)
y = x + np.random.normal(loc = 50, scale = .1, size = 50)

stats.pearsonr(x,y)
```

If you want to set a seed for your random arrays in python you will need to use a different module, random

```python
import random
```

```python
random.seed(1303)
x = [None]*50
for i in np.arange(len(x)):
    x[i] = random.gauss(mu = 0, sigma = 1)
print(x)
```

numpy also provides many of the mean, variance, standard deviation functions described

```python
print(np.mean(x))
print(np.var(x))
print(np.sqrt(np.var(x)))
print(np.std(x))
```

np.arange can be used like seq to get sequences, range, the built in function, also works well

```python
print(range(1, 10))
print(np.arange(1,10))
print(np.arange((-1 * np.pi), (np.pi), (np.pi/25)))
```

```python
x = np.arange((-1 * np.pi), (np.pi), (np.pi/25))
y = x
f = [None]*len(x)
for i in np.arange(len(x)):
    f[i] = np.cos(y[i])/(1 + np.power(x[i], 2))
    
x,y = np.meshgrid(x, y)
```

matplot lib does not have support for 3d plots so I use mplot3d

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(x, y, f, cmap = cm.coolwarm, antialiased = False, linewidth = 0)
```

indexing in pandas!

```python
#you can select a column by name
d['year']
```

```python
#you can select rows by number; note that unlike R, python starts indexing at 0
d[0:10]
```

```python
#you can select both by both
d['year'][0:10]
```

```python
#a difference between pandas and r, is if you flip the name and the numbers you get the same 
#thing (and not an error)
d[0:10]['year']
```

```python
#you can select by label using .loc; this is a little odd here because my indexes are just
#numbers anyway but there you go
d.loc[[0],['year']]
```

```python
#note with loc order matters!
```

```python
#or you can index by position using iloc
d.iloc[[0],[0]]
```

```python
#note for these you can also use slices
print('.loc')
print(d.loc[0:3,['city','year']])
print('')
print('.iloc')
print(d.iloc[0:3, 0:2])

#note loc is inclusive and iloc is NOT inclusive
```

```python
#and you can select by value
d[d.year == 2010]
```
